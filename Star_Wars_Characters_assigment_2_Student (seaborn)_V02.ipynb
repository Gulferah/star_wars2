It builds on top of matplotlib and integrates closely with pandas data structures.

Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">"TASK_1 Plotting the Star Wars characters by genders"</p>

style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
# gender

df['gender'].value_counts()
```

**This can now be plotted much more concisely.**<br>
Axes.bar_label() automatically labels bars [since matplotlib 3.4](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar_label.html)

```python
# with seaborn barplot


ax = sns.countplot(data=df, x="gender", order=df['gender'].value_counts().index)        
for p in ax.patches:
    ax.annotate(round(p.get_height()), (p.get_x()+0.4, p.get_height()+1), ha="center")
    
plt.title('Gender', color= 'blue')
plt.xticks(rotation=45);
```

```python
# with seaborn countplot

ax = sns.countplot(data=df, x="gender", order=df['gender'].value_counts().index)  

for p in ax.patches:
    ax.annotate(round(p.get_height()), (p.get_x()+0.4, p.get_height()+1), ha="center")
    
    plt.title('Gender', color= 'blue')
    plt.xticks(rotation=45, color= 'blue');
```

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">"TASK_2 Plotting the characters by eye color"</p>
 

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
# eye_color
df['eye_color'].value_counts()
```

```python
# with seaborn barplot

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=df['eye_color'].value_counts().index, y=df['eye_color'].value_counts(), order=df['eye_color'].value_counts().index)

for container in ax.containers:
    ax.bar_label(container)
    
plt.title('Eye Color', color='blue')   
plt.xlabel('eye_color')
plt.ylabel('')
plt.xticks(rotation=45);
```

```python
# with seaborn countplot


plt.figure(figsize=(10, 6))
ax= sns.countplot(x='eye_color', data=df, order=df['eye_color'].value_counts().index)       

for container in ax.containers:
    ax.bar_label(container)

plt.title('Eye Color', color='blue')   
plt.xticks(rotation=45, color='blue');

```

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">TASK_3 - Plotting the MASS under 3 groups: "High_mass", "Normal_mass", "Low_mass" </p>


<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
# mass_cat

df['mass_cat'] = pd.cut(df['mass'], 
                        bins=[-float('inf'), 50, 100, float('inf')],
                        labels=['Low_mass', 'Normal_mass', 'High_mass'])

df['mass_cat'].value_counts(dropna = False)
```

```python
# with seaborn barplot

plt.figure(figsize = (8,6))

sns.barplot(x=df['mass_cat'].value_counts().index, y=df['mass_cat'].value_counts(), order=df['mass_cat'].value_counts().index)
plt.title('Mass Groups', fontdict={'color': 'blue', 'fontsize': 14})


for i, count in enumerate(df['mass_cat'].value_counts()):
    plt.text(i, count, str(count), ha='center', va='bottom');

```

**Seaborn color palettes**

sns.color_palette("bone", n_colors=20)


https://seaborn.pydata.org/tutorial/color_palettes.html
https://www.practicalpythonfordatascience.com/ap_seaborn_palette

```python
# with seaborn countplot

plt.figure(figsize=(8, 6))
ax = sns.countplot(x='mass_cat', data=df, palette='Reds', order=['Normal_mass', 'Low_mass', 'High_mass'])

for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', color='red')

plt.title('Mass Groups', color='blue')
plt.xlabel('mass_cat')
plt.xticks(rotation=45);
```

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">"TASK_4 - Plotting the species top 5"</p>


<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
#species_freq

species_freq = df['species'].value_counts().head()
species_freq
```

https://www.statology.org/seaborn-barplot-color/

```python
# with seaborn Barplot

ax = sns.barplot(x=species_freq.index, y=species_freq.values, palette='tab10')

for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', color='black')

plt.title('Species Top 5', color='cyan')
plt.xlabel('species');
```

```python

species_top5 = df['species'].value_counts().head(5)

plt.figure(figsize=(8, 6))

ax = sns.countplot(data=df[df['species'].isin(species_top5.index)], x='species', order=species_top5.index, palette='tab10')
plt.title('Species Top 5', fontdict={'color': 'c', 'fontsize': 14})

for p in ax.patches:
    ax.annotate(round(p.get_height(), 1), (p.get_x() + p.get_width() / 2, p.get_height() - 1.5), ha="center")
    
plt.xticks(rotation=90);

```

```python
# skin_color top 7

df['skin_color'].value_counts().head(7)
```

```python
# with seaborn Barplot


data = df['skin_color'].value_counts().head(7)

ax = sns.barplot(x=data.index, y=data.values)


for i, count in enumerate(data.values):
    ax.text(i, count + 0.1, str(count), ha='center', color='black')


plt.xlabel('skin_color')
plt.ylabel('skin_color', color='red')
plt.xticks(rotation=45)
plt.title('Skin Color Top 7', color='red');
```

```python
# with seaborn countplot


ax= sns.countplot(data=df, x='skin_color', order=df['skin_color'].value_counts().head(7).index)

for i, count in enumerate(data.values):
    ax.text(i, count + 0.1, str(count), ha='center', color='black')
    
plt.xlabel('skin_color')
plt.title('Skin Color Top 7', color='blue');

```

### <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:left; border-radius:10px 10px;">6.1 - Plotting the Box plot of height</p></p>

```python
# with seaborn boxplot

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='height');
```

### <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:left; border-radius:10px 10px;">6.2 - Plotting the Swarm Plot of height in genders separeted by mass_cat.</p>


```python
# with seaborn Swarmplot

plt.figure(figsize=(12, 6))

sns.swarmplot(data=df, x='gender', y='height', hue='mass_cat', palette = {'Normal_mass' : 'blue', 'Low_mass': 'red', 'High_mass': 'green'})
plt.legend(loc='upper right', title='mass_cat');
```
